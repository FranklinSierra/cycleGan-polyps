```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
import os
import shutil
import random
```

# Kfold stratified
Train: 80%, Test: 20%
* Ade: son el 53% de las muestras
* Hyp: son el 28% de las muestras
* Ser: son el 19% de las muestras

```python
ade = ["adenoma_/video_" + str(i) for i in range(1,41)]
hyp = ["hiperplastic_/video_" + str(i) for i in range(1,22)]
ser = ["serrated_/video_" + str(i) for i in range(1,16)]
```

```python
videos = np.concatenate((ade, hyp, ser), axis=0)
videos
```

```python
ade_label = np.zeros(40, dtype = int)
hyp_label = np.ones(21, dtype = int)
ser_label = np.ones(15, dtype = int) + 1
```

```python
labels = np.concatenate((ade_label, hyp_label, ser_label), axis = 0)
skf = StratifiedKFold(n_splits=4, shuffle=True, random_state=10)
```

## Load and saving images into kfold folder...

```python
dic = {0: "adenoma",
      1: "hiperplastic",
      2: "serrated"}
```

```python
#LEYENDO Y GUARDANDO PARA EL CONJUNTO DE TRAIN (AMBAS FUENTES)
random.seed(10)
path = '/Users/frank/Documents/Tesis/Polipos/DataSet/PolypDatasetProcesed/'
save_path = '/Users/frank/Documents/maestria/propuesta/experiments/data/'
c = 0
for train_index, test_index in skf.split(videos, labels):
    c += 1
    print("***KFOLD ***", c)
    #accedo al path general de NBI and WL
    nbi_pth = path + "NBI/"
    wl_pth = path + "WL/"
    
    #tomo el nombre tanto de videos como labels por el indice
    x_train_fold, x_test_fold = videos[train_index], videos[test_index]
    y_train_fold, y_test_fold = labels[train_index], labels[test_index]
    
    #itero sobre train para NBI and WL
    for video, label in zip(x_train_fold, y_train_fold):
        #tomo la clase
        clase = dic[label]
        print("clase: ", clase)
        #armo el path para las dos fuentes
        nbi = video.split('/')[0] + 'NBI/' + video.split('/')[-1]
        wl = video.split('/')[0] + 'WL/' + video.split('/')[-1]
        #path completo para las dos fuentes
        class_nbi_pth = nbi_pth + nbi
        class_wl_pth = wl_pth + wl
        #conozco la cantidad de imgs en cada video 
        can_nbi = len(os.listdir(class_nbi_pth))
        can_wl = len(os.listdir(class_wl_pth))
        
        #cantidad de imgs por clase
        if clase == "adenoma":
            less = 26
        if clase =="hiperplastic":
            less = 48
        else:
            less = 64
        #tomo aleatoriamente la cantidad minima de todas las disp
        imgs_nbi = random.sample(range(1,can_nbi),less)
        imgs_wl = random.sample(range(1,can_wl),less)
        
        #reading and saving nbi images 
        for img in imgs_nbi:
            print("video ", video.split('/')[-1], "...")
            original_pth_nbi = class_nbi_pth + '/' + clase + '_NBI_' + video.split('/')[-1] + '_img_' + str(img) + '.png'
            target_pth = save_path + 'fold' + str(c) + '/train_NBI/'
            shutil.copy(original_pth_nbi, target_pth)
            print("guardado con exito!")
        #reading and saving wl images
        for img in imgs_wl:
            print("video ", video.split('/')[-1], "...")
            original_pth_wl = class_wl_pth + '/' + clase + '_WL_' + video.split('/')[-1] + '_img_' + str(img) + '.png'
            target_pth = save_path + 'fold' + str(c) + '/train_WL/'
            shutil.copy(original_pth_wl, target_pth)
            print("guardado con exito!")       
            
            
```
